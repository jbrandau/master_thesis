```python
def concatDataTraces(path):
    mats = []
    index = 0 
    fileNames = np.arange(100, (d_len + 100), 100).tolist()
    for i in tnrange(0,num_files):
        mats = sio.loadmat(path+"traces_"+ str(fileNames[index]) +".mat", verify_compressed_data_integrity=False)
        mats =  np.array(mats['data'])
        mats =  mats.astype('uint8')
        sio.savemat('../real_traces/SBOX_traces2019-12-24_10_54_27/spot_4_12/sliced/traces_'+str(fileNames[index])+'.mat', {'data': mats}) 
        index = index + 1
        mats = []
```
