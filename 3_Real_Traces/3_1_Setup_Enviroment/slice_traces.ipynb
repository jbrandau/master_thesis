```python
fixedDir = './SBOX_traces2019-10-31_17_08_37/' 
newDir = fixedDir + 'traces/'

for i in tnrange(0,17):
    for j in range(0,20):
        df_traces = pd.read_csv(newDir + 'trace_' + str(i) + '_' + str(j) + '.csv',low_memory=False)
        tmpSliced = df_traces.transpose()[850:2150].copy().reset_index(drop=True).transpose()
        tmpSliced.to_csv(newDir + "trace_sliced_" + str(i) + "_" + str(j) + ".csv",index=False)
```
