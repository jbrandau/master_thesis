```python
import os
import scipy.io as sio
import pandas as pd

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
dataDir = "./SBOX_traces2019-10-31_17_08_37/spot_0_0/traces/"
mats = []
for file in os.listdir( dataDir ) :
    mats.append( sio.loadmat( dataDir+file ) )
```

```python
def concatData(data):
    myArray = []
    for i in range(0,10):
        for j in range(0,1000):
            myArray.append(data[i]['data'][j])
    return(myArray)
```

```python
df_data1 = pd.DataFrame(concatData(mats))
```

```python
df_data1[:1].transpose().plot()
df_data1[:2].transpose().plot()
df_data1[:3].transpose().plot()
```

```python
df_data1['label'] = 123
```

```python
df_data1.to_csv("./fixed_123_2500/fix123.csv",index=False)
```

```python
dataDir = "./random_2500/traces/"
mats = []
for file in os.listdir( dataDir ) :
    mats.append( sio.loadmat( dataDir+file ) )
```

```python
df_data3 = pd.DataFrame(concatData(mats))
```

```python
dataDir = "./random_2500/labels/"
matsL = []
for file in os.listdir( dataDir ) :
    matsL.append( sio.loadmat( dataDir+file ) )
```

```python
def concatDataL(data):
    myArray = []
    for i in range(0,20):
        for j in range(0,500):
            
            try:
                myArray.append(data[i]['nonces'][j][0])
            except:
                myArray.append(data[i]['noncesA'][j][0])
    return(myArray)
```

```python
df_data3['label'] = concatDataL(matsL)
```
